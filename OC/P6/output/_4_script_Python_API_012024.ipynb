```python
import requests
import pandas as pd
import numpy as np
import json  

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {"nutrition-type":"cooking"}

headers = {
    'X-RapidAPI-Key': '928542342emsh27446b85a20c98cp15dc92jsnc6203c2b3c0e',
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)  
data = response.json()

with open("./api_response.json", 'w') as file:  
    json.dump(data, file) 
# # Cr√©ation d'un dataframe pour les produits  
# df = pd.DataFrame(products)  
# df = df[["food.foodId", "food.label", "food.category", "food.foodContentsLabel", "food.image"]]  
# df.columns = ["foodId", "label", "category", "foodContentsLabel", "image"]  
```

```python
products = []  
for hint in data["hints"][:10]:  
    food = hint["food"]  
    product = {  
        "foodId": food["foodId"],  
        "label": food["label"],  
        "category": food["category"],  
        "foodContentsLabel": ";".join([f"{key}: {value}" for key, value in food["nutrients"].items()]),
        "image": food.get("image", np.nan)
    }  
    products.append(product)  
  
df = pd.DataFrame(products)  
# df.to_csv("products.csv", index=False)  
```
