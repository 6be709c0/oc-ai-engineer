```python
from helpers import *

# Loop between periods ([1 month, 2 month, ..., 12 months]) (For all 3 years)
df = pd.read_csv("full_data_undersample.csv").drop(columns=["customer_unique_id"], axis=1)
df['last_order_date'] = pd.to_datetime(df['last_order_date'])  # Group by year and month  

# For each period
#   - Calculate cluster of all clients

#   - Dataframe should look like
#   Period      Cluster
#   10/2016     4

#   - Calculate the difference of cluster
#   df["has_changed"] = first cluster vs all
```

```python
df.head()
```

```python
# Calculate the periods
pca = PCA(n_components=2)  
df_pca = pca.fit_transform(df.drop(columns=["last_order_date"], axis=1))

  # Fit Kmeans model  
kmeans = KMeans(n_clusters=10, n_init="auto")  
kmeans.fit(df_pca) 

# Predict cluster labels  
labels = kmeans.predict(df_pca)
df['cluster'] = labels

df['cluster'].value_counts()
```
