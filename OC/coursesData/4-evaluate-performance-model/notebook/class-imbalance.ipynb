```python
import numpy as np  
import pandas as pd  
from sklearn.utils.class_weight import compute_class_weight
# Assuming two features X1 and X2  
n_samples = 1000  
X1_idle = np.random.normal(loc=5, scale=1, size=int(n_samples * 0.9))
X2_idle = np.random.normal(loc=5, scale=1, size=int(n_samples * 0.9))

X1_wave = np.random.normal(loc=10, scale=1, size=int(n_samples * 0.1))
X2_wave = np.random.normal(loc=10, scale=1, size=int(n_samples * 0.1))

data_idle = pd.DataFrame({'X1': X1_idle, 'X2': X2_idle, 'Class': 'idle'})  
data_wave = pd.DataFrame({'X1': X1_wave, 'X2': X2_wave, 'Class': 'wave'})  

dataset = pd.concat([data_idle, data_wave], axis=0).reset_index(drop=True)
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  

sns.scatterplot(data=dataset, x='X1', y='X2', hue='Class')  
plt.xlabel('X1')  
plt.ylabel('X2')  
plt.title('Dataset with Class Imbalance')  
plt.show()  
```

```python
class_counts = dataset['Class'].value_counts()  
class_labels = class_counts.index  
class_values = class_counts.values  
  
sns.barplot(x=class_labels, y=class_values)  
plt.xlabel('Class')  
plt.ylabel('Number of Samples')  
plt.title('Class Imbalance')  
plt.show()  
```
