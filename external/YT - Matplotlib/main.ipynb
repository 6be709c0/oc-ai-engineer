```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y, label="2x", color="red",linewidth=5, linestyle="--", marker=".", markersize=12, markerfacecolor="green", markeredgecolor="green")

plt.title("My title", fontdict={'fontname':"Comic Sans MS", "fontsize": 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# Short format
plt.plot(x,y, "r.--", label="2x")

plt.title("My title", fontdict={'fontname':"Comic Sans MS", "fontsize": 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# Short format
plt.plot(x,y, "r.--", label="2x")

x2 = np.arange(0,4.5,0.5)
print(x2, x2**2)
plt.plot(x2,x2**2, label="x^2")

plt.title("My title", fontdict={'fontname':"Comic Sans MS", "fontsize": 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# Short format
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x,y, "r.--", label="2x")

x2 = np.arange(0,4.5,0.5)
print(x2, x2**2)
plt.plot(x2[:6],x2[:6]**2, label="x^2")
plt.plot(x2[5:],x2[5:]**2, 'r--')

plt.title("My title", fontdict={'fontname':"Comic Sans MS", "fontsize": 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.savefig("mygraph.png", dpi=300)
plt.show()
```
